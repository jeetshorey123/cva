## Q2-a) 
**Describe Kruskal's Algorithm and Union-Find Data Structure Used for Cycle Detection in MST**  
- **Kruskal's Algorithm**: A greedy algorithm to find the Minimum Spanning Tree (MST) of a graph. It sorts all edges by weight and adds them to the MST in order of increasing weight, provided they do not form a cycle.  
- **Union-Find Data Structure**:  
  - Used to detect cycles efficiently.  
  - Maintains a set of disjoint subsets to track connected components.  
  - Operations:  
    1. **Find**: Determines which subset a particular element belongs to.  
    2. **Union**: Merges two subsets.  
    3. **Path Compression**: Optimizes the find operation.

## Q4-a)
**Difference Between 2-3 Tree, B-Tree, and Red-Black Tree**
- **2-3 Tree**: Each node has 2 or 3 children. Maintains strict balancing.
- **B-Tree**: A generalization of 2-3 trees. Optimized for disk storage with multiple keys per node.
- **Red-Black Tree**: A binary search tree with color attributes for balancing.
